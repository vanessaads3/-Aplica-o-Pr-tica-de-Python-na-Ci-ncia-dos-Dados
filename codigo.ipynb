````python
#Ver tipos de dados
tipos = df.dtypes
print(tipos)
````

````python
#Separar colunas numéricas
numericas = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print("Características numéricas:", numericas)
````

````python
#Separar colunas categóricas
cateoricas = df.select_dtypes(include=['object']).columns.tolist()
print("Características categóricas:", cateoricas)
````

````python
df[numericas].describe()
````

````python
for col in cateoricas:
  print(f"\nFrequência de valores em '{col}':")
  print(df[col].value_counts())
````

````python
faltantes = df.isnull().sum()
print("Dados faltantes por coluna:")
print(faltantes[faltantes > 0])
````
