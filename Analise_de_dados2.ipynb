### Exercício 2: Diferença entre distribuições

As distribuições das características de valor de fatura e de pagamento apresentam diferenças visíveis. Os valores de fatura (`BILL_AMT`) têm distribuições mais contínuas e amplas, com maior variabilidade e caudas longas, indicando que alguns clientes acumulam dívidas elevadas. Já os valores de pagamento (`PAY_AMT`) são mais discretos, com picos em valores específicos e uma concentração significativa em zero, sugerindo que muitos clientes não realizaram pagamentos em determinados meses. Essa diferença pode refletir comportamentos distintos entre consumo e quitação de dívidas.

```python
# Criar os bins
bin = pd.cut(df['BILL_AMT1'], bins=20)

# Calcular a média de pagamento para cada bin
bin_means = df.groupby(bin)['PAY_AMT1'].mean()

# Extrair os centros dos bins e os valores médios
bin_centers = [interval.mid for interval in bin_means.index]
X = bin_centers
y = bin_means.values
```

```python
# Importa as bibliotecas necessárias
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Divide os valores da coluna 'BILL_AMT1' em 20 faixas (intervalos)
bins = pd.cut(df['BILL_AMT1'], bins=20)

# Calcula a média dos valores da coluna 'PAY_AMT1' para cada faixa de 'BILL_AMT1'
bin_means = df.groupby(bins)['PAY_AMT1'].mean()

# Extrai o ponto médio de cada faixa para usar como variável independente (X)
bin_centers = [interval.mid for interval in bin_means.index]

# Converte os centros das faixas e as médias em arrays NumPy
X = np.array(bin_centers).reshape(-1, 1)  # X: centros das faixas (fatura)
y = np.array(bin_means.values)           # y: média de pagamento por faixa

# Remove valores nulos (NaN) dos dados para evitar erros no modelo
mask = ~np.isnan(y)
X_clean = X[mask]
y_clean = y[mask]

# Cria e treina o modelo de regressão linear com os dados limpos
model = LinearRegression()
model.fit(X_clean, y_clean)

# Define uma função que estima o valor de pagamento com base na fatura
def estimar_pagamento(valor_fatura):
    return model.predict(np.array([[valor_fatura]]))[0]
```

```python
# Gera as previsões de pagamento usando o modelo treinado
y_pred = model.predict(X_clean)

# Cria um histograma dos valores reais de pagamento
plt.hist(y_clean, bins=20, alpha=0.5, label='Real')

# Cria um histograma dos valores estimados pelo modelo
plt.hist(y_pred, bins=20, alpha=0.5, label='Estimado')

# Adiciona a legenda para identificar os dois histogramas
plt.legend()

# Define o título do gráfico
plt.title('Comparação de Pagamentos')

# Define o rótulo do eixo X
plt.xlabel('Valor do Pagamento')

# Define o rótulo do eixo Y
plt.ylabel('Frequência')

# Adiciona uma grade ao fundo do gráfico para facilitar a leitura
plt.grid(True)

# Ajusta automaticamente o layout para evitar sobreposição de elementos
plt.tight_layout()

# Exibe o gráfico
plt.show()
```

```python
from sklearn.model_selection import train_test_split


# Seleciona as colunas de valores de fatura como variáveis independentes (X)
X_bill = df[['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']]

# Define a variável dependente (y) como o valor de pagamento da primeira parcela
y_pay = df['PAY_AMT1']

# Divide os dados em conjunto de treino (80%) e teste (20%)
X_train, X_test, y_train, y_test = train_test_split(X_bill, y_pay, test_size=0.2, random_state=42)

# Cria o modelo de regressão linear múltipla
model_multi = LinearRegression()

# Treina o modelo com os dados de treino
model_multi.fit(X_train, y_train)

# Faz previsões com os dados de teste
y_pred_multi = model_multi.predict(X_test)

# Plota histogramas comparando os valores reais e estimados de pagamento
plt.hist(y_test, bins=30, alpha=0.5, label='Real')         # Valores reais
plt.hist(y_pred_multi, bins=30, alpha=0.5, label='Estimado') # Valores previstos
plt.legend()
plt.title('Regressão Múltipla com BILL_AMT')
plt.show()

# Seleciona todas as variáveis financeiras como preditoras
X_all = df[financeiras]

# Cria um novo modelo de regressão linear
model_all = LinearRegression()

# Treina o modelo com todas as variáveis financeiras
model_all.fit(X_all, y_pay)

# Faz previsões com o próprio conjunto de dados
y_pred_all = model_all.predict(X_all)

# Plota histogramas comparando os valores reais e estimados
plt.hist(y_pay, bins=30, alpha=0.5, label='Real')           # Valores reais
plt.hist(y_pred_all, bins=30, alpha=0.5, label='Estimado')  # Valores previstos
plt.legend()
plt.title('Regressão Múltipla com Todas as Características')
plt.show()


```
